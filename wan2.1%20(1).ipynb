```python
import os
from IPython.display import Video, display

# Specify the directory path (change this to your directory)
directory = "./"  # Current directory, or replace with your path like "C:/Videos/"

# Get all .mp4 files in the directory
mp4_files = [f for f in os.listdir(directory) if f.endswith('.mp4')]

# Print the list of found files
print(f"Found {len(mp4_files)} MP4 files:")
for i, file in enumerate(mp4_files, 1):
    print(f"{i}. {file}")

# Display each video
for mp4_file in mp4_files:
    video_path = os.path.join(directory, mp4_file)
    print(f"\nPlaying: {mp4_file}")
    display(Video(video_path, embed=True, width=500))  # Adjust width as needed
```
